# Fuel Mix 

Grabs Fuel Mix from gridiso.com 

```python
#import modules 
import gridstatus as gs 
import pandas as pd
import os
import pathlib
from glob import glob

```

```python
#Create data directory 
data_dir = os.path.join(
    pathlib.Path.home(),
    'code-projects',
    'generation-analysis',
    )
os.makedirs(data_dir, exist_ok=True)

fuel_dir = os.path.join(data_dir, 'fuel-mix')
os.makedirs(fuel_dir, exist_ok=True)

fuel_data = os.path.join(fuel_dir, 'caiso-fuel-mix.csv')
```

```python

caiso=gs.CAISO()

#set dates
start = pd.Timestamp("April 10, 2018").normalize()
end = pd.Timestamp.now().normalize()


if not glob(fuel_data):
    mix_df = caiso.get_fuel_mix(start, end=end, verbose=False)
    ca_mix_df = mix_df.to_csv(fuel_data)
else:
    ca_mix_df = pd.read_csv(fuel_data)


ca_mix_df    
```
